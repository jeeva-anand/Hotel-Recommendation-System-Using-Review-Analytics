```python
cards = soup.find_all("li", class_="review_item")

reviews = []
dates = []
names = []
reviews_data = []

for item in cards:
    review = item.find("span", itemprop="reviewBody")
    date = item.find("p", class_="review_staydate")
    name = item.find("span", itemprop="name")

    if review:
      reviews.append(review.get_text(strip=True) if review else None)

    if date:
        dates.append(date.get_text(strip=True) if date else None)

    if name:
          names.append(name.get_text(strip=True) if name else None)

    reviews_data.append({
        "name": name.get_text(strip=True) if name else None,
        "review": review.get_text(strip=True) if review else None,
        "stay_date": date.get_text(strip=True) if date else None
    })

reviews_data
```
